Prove Lemma 2.1 in [The Epoch-Greedy Algorithm for Contextual Multi-armed Bandits](https://papers.nips.cc/paper_files/paper/2007/file/4b04a686b0ad13dce35fa99fa4161c65-Paper.pdf)

\begin{align*}
\mathbb{E}_{\mathbf{r} \sim P|\mathbf{x}, a \sim Q} [\hat{r}_{a'}(r_a)] = \mathbb{E}_{\mathbf{r} \sim P|\mathbf{x}} \left[ r_{a'} \right ]
\end{align*}

where

$$
\hat{r}_{a'}(r_a) = \frac{1}{q(a)} r_a \mathbb{1}[a' = a]
$$

Note, 

* $P|\mathbf{x}$ is the conditional distribution of rewards vector $\mathbf{r}$ given context $\mathbf{x}$.
* $Q$ is a probability distribution over all actions.
* $r_{a}$ is the observed reward after action $a$ is pulled.
* $\hat{r}_{a'}(r_a)$ is the estimated reward of action $a'$ based on the observed reward of action $a$.
* $q(a)$ is the probability of pulling action $a$ based on $Q$. 

Proof

\begin{align*}
\mathbb{E}_{\mathbf{r} \sim P|\mathbf{x}, a \sim Q} [\hat{r}_{a'}(r_a)]
&= \mathbb{E}_{\mathbf{r} \sim P|\mathbf{x}} \left[ \sum_{a=1}^k \hat{r}_{a'}(r_a) q(a) \right ] \\
&= \mathbb{E}_{\mathbf{r} \sim P|\mathbf{x}} \left[ \sum_{a=1}^k \frac{1}{q(a)} r_a \mathbb{1}[a' = a] q(a) \right ] \\
&= \mathbb{E}_{\mathbf{r} \sim P|\mathbf{x}} \left[ \sum_{a=1}^k r_a \mathbb{1}[a' = a] \right ] \\
&= \mathbb{E}_{\mathbf{r} \sim P|\mathbf{x}} \left[ r_{a'} \right ]
\end{align*}

Consider the case, we're using a hypothesis $h(x)$ that maps a context to an action, so we're interested in $r_{h(\mathbf{x})}$. Using Lemma 2.1, we can estimate it with 

\begin{align*}
\hat{r}_{h(\mathbf{x})}(r_a) = \frac{1}{q(a)} \mathbb{1} \left(h(\mathbf{x}) = a \right) r_a
\end{align*}

and

\begin{align*}
\mathbb{E}_{\mathbf{r} \sim P|\mathbf{x}, a \sim Q} [\hat{r}_{h(\mathbf{x})}(r_a)] 
&= \mathbb{E}_{\mathbf{r} \sim P|\mathbf{x}, a \sim Q} \left[\frac{1}{q(a)} \mathbb{1} \left(h(\mathbf{x}) = a \right) r_a \right]  \\
&= \mathbb{E}_{\mathbf{r} \sim P|\mathbf{x}} \left[ r_{h(\mathbf{x})} \right ]
\end{align*}
