By Beta distribution, the mean is 

\begin{align*}
\mu(B_a(t - 1)) &= \frac{\alpha + \sum_{s=1}^{t - 1} \mathbb{1}(A_t = a, R_t=1) }{\alpha + \beta + \sum_{s=1}^{t-1} \mathbb{1}(A_t = a)}
\end{align*}

Note,

* $\sum_{s=1}^{t - 1} \mathbb{1}(A_t = a, R_t = 1)$ is the number of wins for playing arm $a$.
* $\sum_{s=1}^{t-1} \mathbb{1}(A_t = a)$ is the number of plays of arm $a$.

Linear contextual bandit:

* At round $t$, observe $C_t \in \mathcal{C}$, where $\mathcal{C}$ is the set of all possible contexts.
* Standard idea, design feature mapping: $\phi: [k] \times C \rightarrow \mathbb{R}^d$, where $k$ is the number of arms.
* Assume reward $R_{t, a} = \langle \phi(a, C_t), \theta \rangle + \eta_t$ for some unknown $\theta \in \mathbb{R}^d$.

* **Simplified view**: The learners receives action set: $\mathcal{A}_t = \{ \phi(a, C_t): a \in [k]\} \subset \mathbb{R}^d$

* Plays action $A_t \in \mathcal{A}_t \subset \mathbb{R}^d$
* Reward is $\langle A_t, \theta \rangle + \eta_t$
* Regret is

\begin{align*}
\Re_n 
&= \sum_{t = 1}^n \langle a^* - A_t, \theta \rangle \\
&= \sum_{t = 1}^n \max_{a \in \mathcal{A}_t} \langle a - A_t, \theta \rangle
\end{align*}

Note, $a^* = \arg \max_{a \in \mathcal{A}_t} \langle a , \theta \rangle$
