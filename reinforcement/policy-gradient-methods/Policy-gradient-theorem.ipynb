# Trying to understand policy gradient theorem at Page 339:

\begin{align*}
\nabla v_{\pi} (s) &= \sum_{a}\bigg [ \nabla_{\pi}(a|s) q_{\pi}(s, a) + \pi(a|s) \gamma \sum_{s'} p(s'|s, a) \nabla v_{\pi}(s') \bigg ]
\\
\nabla v_{\pi} (s') &= \sum_{a'}\bigg [ \nabla_{\pi}(a'|s') q_{\pi}(s', a') + \pi(a'|s') \gamma \sum_{s'} p(s''|s', a') \nabla v_{\pi}(s'') \bigg ]
\\
\nabla v_{\pi} (s'') &= \sum_{a'}\bigg [ \nabla_{\pi}(a''|s'') q_{\pi}(s'', a'') + \pi(a''|s'') \gamma \sum_{s''} p(s'''|s'', a'') \nabla v_{\pi}(s''') \bigg ]
\end{align*}
