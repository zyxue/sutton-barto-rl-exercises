```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
```

# Demo of `np.meshgrid`

```python
x1s=np.arange(0, 10, 3)
x2s=np.arange(0, 80, 30)
```

```python
x1s_meshed, x2s_meshed = np.meshgrid(x1s, x2s)
```

```python
x1s_meshed, x2s_meshed
```

```python
x1s_meshed_ravel = x1s_meshed.ravel()
x2s_meshed_ravel = x2s_meshed.ravel()
```

```python
x1s_meshed_ravel
```

```python
x2s_meshed_ravel
```

```python
zs = x1s_meshed ** 2 + x2s_meshed
```

```python
zs.reshape(x1s_meshed.shape)
```

```python
plt.contour(x1s_meshed, x2s_meshed, zs)
plt.xlabel('x1')
plt.xlabel('x2')
```

# `np.mgrid` output appears to match that of `np.meshgrid` when `indexing='ij'`

```python
meshgrid_output = np.meshgrid(x1s, x2s, indexing='ij')
```

```python
mgrid_output = np.mgrid[
    0:10:3,
    0:80:30
]
```

```python
(meshgrid_output == mgrid_output).all()
```
