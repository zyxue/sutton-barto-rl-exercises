Currently, just visualization of huber loss is shown.

```python
import itertools
import pickle
import math

import altair as alt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

\begin{align*}
L = \begin{cases}
  \frac{1}{2} (y - t)^2              & \text{if $|y - t| \le \alpha$} \\
  \alpha(|y - t| - \frac{\alpha}{2}) & \text{if $|y - t| > \alpha$}
\end{cases}
\end{align*}

```python
def huber(deltas, alpha):
    return np.where(np.abs(deltas) <= alpha, 0.5 * deltas ** 2, alpha * (np.abs(deltas) - alpha / 2))
```

```python
df_delta = pd.Series(np.arange(-10, 10, 0.1)).to_frame(name='delta')
```

```python
dfs = []
for alpha in [0, 1, 3, 6, 11]:
    _df = df_delta.assign(loss=lambda df: huber(df.delta, alpha), alpha=alpha)
    dfs.append(_df)
ndf = pd.concat(dfs)
```

```python
ndf.head(2)
```

```python
alt.Chart(ndf).mark_line(size=4).encode(x='delta', y='loss', color='alpha',
#                                         strokeWidth=alt.value(4),
                                       )
```
