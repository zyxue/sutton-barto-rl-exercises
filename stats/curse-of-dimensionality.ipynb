```python
import matplotlib.pyplot as plt
import numpy as np
```

To show the volume changes in high-dimension space:

\begin{align*}
V_D(r) &= K_D r^D \\
\frac{V_D(1) - V_D(1 - \epsilon)}{V_D(1)} &= \frac{K_D 1^D - K_D(1 - \epsilon)^D}{K_D 1^D} \\
&= 1 - (1-\epsilon)^D
\end{align*}

```python
def volume_fraction(epsilon, D):
    return 1 - (1 - epsilon) ** D
```

```python
epsilons = np.arange(0, 1.02, 0.02)
```

```python
for D in [1, 2, 3, 10, 20, 50]:
    plt.plot(epsilons, volume_fraction(epsilons, D), label=f"{D}D")

plt.plot([0.1, 0.1], [0, 1], label=f"ϵ=0.1", linestyle='--', lw=1, color='black')

plt.legend()
plt.grid()
plt.xlabel('ϵ')
plt.ylabel('Volume fraction')
```

This plot shows that in 50D space, for a unit hypersphere (49-sphere), even when $\epsilon$ is as small as 0.1, the volume in the thin shell of between 1 and 1 - $\epsilon$ has already almost taken the whole volume of the sphere, which can be very counter-intuitive from a low-dimension-space perspective that we are more used to.
